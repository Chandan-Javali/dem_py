# **MATPLOTLIB**

```python
# import matplot lib, numpy, xpoint is np of array of 0,6 , y point is 0,256 , then plot x and ypoint

import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([0, 6])
y_points = np.array([0, 256])

plt.plot(x_points, y_points)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot of X and Y points")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define y_points
y_points = np.array([3, 8, 10, 14])

# Generate x_points with the same length as y_points
x_points = np.arange(len(y_points)) # x_points will be [0, 1, 2, 3]

# Plot the data
plt.plot(x_points, y_points, marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot of X and Y points")
plt.show()
```
